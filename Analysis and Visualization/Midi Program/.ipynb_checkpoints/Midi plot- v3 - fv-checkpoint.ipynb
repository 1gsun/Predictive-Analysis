```python
from collections import OrderedDict
fig = plt.figure(figsize = (10, 8))

time_dict = dict()
for i in list(set(df1.Channel)):
    dloop = df1[df1.Channel == i]
    dloop_zero = dloop[dloop.Velocity != 0]
    time_dict[i] = dloop_zero["Real_time"].sum()    
time_dict_sorted = sorted(time_dict.items(), key=lambda x: x[1],reverse= True)

midi_length = max(list(set(df1["Length"]))) # supposed to only have 1 length each song. using max to avoid any problem 
time_list_sorted = list()
channel_list_sorted = list()
pct_list = list()

for i in range(len(time_dict_sorted)):
    channel_list_sorted.append(time_dict_sorted[i][0])
    time_list_sorted.append(time_dict_sorted[i][1])
    pct_list.append(time_dict_sorted[i][1] / midi_length)
    
plt.bar(range(len(time_list_sorted)),time_list_sorted, align="center")

plt.xticks(range(len(time_list_sorted)),channel_list_sorted)

plt.title("Time Presence by Channel")
plt.xlabel("Channel")
plt.ylabel("Sum of Time    (Percentage of Total)")
#plt.xticks(range(len(channel_list)+1),instrutment_list,rotation = 45)

plt.ylim(ymin=min(time_list_sorted)-20)

for a,b,c in zip(range(len(time_list_sorted)),time_list_sorted,pct_list):
    plt.text(a-0.4,b+1.2, "{0:.0f}%".format(c*100))

plt.show()


# global list 
if int(len(channel_list_sorted) * 0.2) < 4:
    
    global_channel_list =  channel_list_sorted[0:4]
else:
    global_channel_list =  channel_list_sorted[0:int(len(channel_list_sorted) * 0.2)+1]
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("Midi_Database_simplified_v4.csv")
df = df.drop("Unnamed: 0",axis = 1)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for c, z in zip(['r','g'], [20, 10]):
    Jazz_vel = df[df['Genre']=='Jazz']['Velocity']
    Classical_vel = df[df['Genre']=='Classical']['Velocity']
    xs = np.random.choice(Jazz_vel,500)
    #randomly choice 500 attribute in the velocity column in Jazz or Classical category distinctly
    #to compare the velocity pattern of these two genres.
    ys = np.random.choice(Classical_vel,500)
    r = [c] * len(xs)
    #cs[0] = 'c'
    ax.bar(xs, ys, zs=z, zdir='y', color=r, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
